```python
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image1 = cv2.imread("badimage.tif")
```

```python
RGB_img = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB_img)
```

```python
b = image1.copy()
# set green and red channels to 0
Red = b[:, :, 0]
Green = b[:, :, 1]
Blue = b[:, :, 2]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle('Horizontally stacked subplots')
ax1.imshow(Red)
ax1.set_title('Red channel')
ax2.imshow(Green)
ax2.set_title('Green channel')
ax3.imshow(Blue)
ax3.set_title('Blue channel')
```

# We can see that Green channel reduces the fluorescent light but is not sufficent to get a good image without fluorescence

```python
plt.imshow(Green)
```

## FILTERING STEP -1

## As we can see that the fluorescent light marks specific colour range. We have exprimented with the image to find out the limit of these range of colors represented by the fluorescent light  and replaced those colors to the brach color 




```python
b1 = RGB_img.copy()
for i in np.arange(0,512,1):
    for j in np.arange(0,512,1):
        if RGB_img[:,:,0][i][j] > 180:
            if RGB_img[:,:,1][i][j] > 100:
                if RGB_img[:,:,2][i][j] < 255:
                    b1[:,:,0][i][j] = 4
                    b1[:,:,1][i][j] = 33
                    b1[:,:,2][i][j] = 48
                    
#                     b1[:,:,0][i][j] = 127
#                     b1[:,:,1][i][j] = 127
#                     b1[:,:,2][i][j] = 127
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(b1)
ax2.imshow(RGB_img)
```

## FILTERING STEP -2 
#### Change the channel to Green

```python
Green = b1[:, :, 1]
```
